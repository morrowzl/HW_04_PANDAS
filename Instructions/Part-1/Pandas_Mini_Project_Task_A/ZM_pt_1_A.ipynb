```python
# Dependencies
import os
import pandas as pd
import csv
```

```python
# load CSV
survey_data = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
survey_df = pd.read_csv(survey_data, encoding='iso-8859-1', low_memory=False)
```

```python
# Read with pandas
survey_df.head()
```

```python
# Inspect all columns
survey_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced = survey_df.iloc[:, [
    0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced = reduced.replace({0.0:"No",1.0:"Yes"})
#reduced["AttendedBootcamp"] = np.where(survey_df2["AttendedBootcamp"] == 0.0, "No", survey_df2["AttendedBootcamp"])
#reduced["AttendedBootcamp"] = np.where(survey_df2["AttendedBootcamp"] == 1.0, "Yes", survey_df2["AttendedBootcamp"])
reduced.head()
```

```python
# Calculate total number of respondents in survey
respondents = len(reduced.index)
respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
campers = reduced.loc[reduced["AttendedBootcamp"] == "Yes", :]
campers.head()
```

```python
# Calculate average age of attendees
ave_age = campers["Age"].mean()
#ave_age = np.round(ave_age, decimals=2)

# Calculate how many people attended a bootcamp
camperscount = len(campers.index)

# Calculate how many attendees hold degrees
degreeholders = campers["SchoolMajor"].count()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
camp_agender = len(campers.loc[reduced["Gender"] == "agender", :])
camp_female = len(campers.loc[reduced["Gender"] == "female", :])
camp_genqueer = len(campers.loc[reduced["Gender"] == "genderqueer", :])
camp_male = len(campers.loc[reduced["Gender"] == "male", :])
camp_trans = len(campers.loc[reduced["Gender"] == "trans", :])
camp_noncisbinary = camperscount - camp_male - camp_female
p100agenderc= camp_agender / camperscount * 100
p100femalec= camp_female / camperscount * 100
p100genqueerc= camp_genqueer / camperscount * 100
p100malec= camp_male / camperscount * 100
p100transc= camp_trans / camperscount * 100
p100noncisbinaryc = camp_noncisbinary / camperscount * 100

# Calculate percentage of respondents who attended a bootcamp
p100campresp = camperscount / respondents * 100
#p100campresp = np.round(p100campresp, decimals=2)

# Calculate percentage of respondents belonging to each gender
resp_agender = len(reduced.loc[reduced["Gender"] == "agender", :])
resp_female = len(reduced.loc[reduced["Gender"] == "female", :])
resp_genqueer = len(reduced.loc[reduced["Gender"] == "genderqueer", :])
resp_male = len(reduced.loc[reduced["Gender"] == "male", :])
resp_trans = len(reduced.loc[reduced["Gender"] == "trans", :])
p100agender = resp_agender / respondents * 100
p100female = resp_female / respondents * 100
p100genqueer = resp_genqueer / respondents * 100
p100male = resp_male / respondents * 100
p100trans = resp_trans / respondents * 100

# Calculate percentage of attendees with a school degree
p100degreecampers = degreeholders / camperscount * 100
#p100degreecampers = np.round(p100degreecampers, decimals=2)

# Calculate average post-bootcamp salary
ave_sal = campers["BootcampPostSalary"].mean()


print(f"Average age: {ave_age}")
print(f"Bootcamp attendees: {camperscount}")
print(f"Total campers w/ degrees: {degreeholders}")
print(f"agender campers: {camp_agender}")
print(f"female campers: {camp_female}")
print(f"genderqueer campers: {camp_genqueer}")
print(f"male campers: {camp_male}")
print(f"trans campers: {camp_trans}")
print(f"% non-cis-binary campers: {p100noncisbinaryc}")
print(f"% respondents to attend bootcamp: {p100campresp}")
print(f"% respondents agender: {resp_agender}")
print(f"% respondents female: {resp_female}")
print(f"% respondents genderqueer: {resp_genqueer}")
print(f"% respondents male: {resp_male}")
print(f"% respondents trans: {resp_trans}")
print(f"% campers w/ degrees: {p100degreecampers}")
print(ave_sal)
```

```python
# Create a new table consolodating above calculations
sum_table = pd.DataFrame({"Respondents (total)":[respondents],
                         "Respondents Attended Bootcamp (total)":[camperscount],
                         "Respondents AND Campers (%)":[p100campresp],
                         "Average Age of Campers (years)":[ave_age], 
                         "Campers w/ Degree (total)":[degreeholders],
                         "Campers w/ Degree (%)":[p100degreecampers], 
                          "Average Salary Post-Bootcamp":[ave_sal], 
#                          "Agender Respondents (total)":[resp_agender], 
#                          "Agender Respondents (%)":[p100agender], 
#                         "Agender Campers (total)":[camp_agender], 
#                         "Agender Campers (%)":[p100agenderc], 
#                         "Female Respondents (total)":[resp_female], 
#                         "Female Respondents (%)":[p100female], 
#                         "Female Campers (total)":[camp_female], 
                         "Female Campers (%)":[p100femalec], 
#                         "Genderqueer Respondents (total)":[resp_genqueer], 
#                         "Genderqueer Respondents (%)":[p100genqueer], 
#                         "Genderqueer Campers (total)":[camp_genqueer], 
#                         "Genderqueer Campers (%)":[p100genqueerc], 
#                         "Male Respondents (total)":[resp_male], 
#                         "Male Respondents (%)":[p100male], 
#                         "Male Campers (total)":[camp_male], 
                         "Male Campers (%)":[p100malec],
                          "Non-cis-binqary Campers (%)":[p100noncisbinaryc]})
#                         "Trans Respondents (total)":[resp_trans], 
#                         "Trans Respondents (%)":[p100trans], 
#                         "Trans Campers (total)":[camp_trans], 
#                         "Trans Campers (%)":[p100transc]})
                            
                          
sum_table
```
