```python
# Dependencies
import pandas as pd
import os
import csv
```

```python
# load CSV
file= os.path.join("Resources/2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas
new_coders_df=pd.read_csv(file, encoding='iso-8859-1', low_memory=False)
```

```python
# Inspect all columns
new_coders_df.head()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_coders_df=new_coders_df.iloc[:,[0,1,2,3,4,7,8,9,10,11,29,30,32,36,37,45,48,56,110,111]]
reduced_coders_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
reduced_coders_df= reduced_coders_df.replace({0.0:"No",1.0:"Yes"})
reduced_coders_df.head()
```

```python
# Calculate total number of respondents in survey
total_surveyed=len(reduced_coders_df)
total_surveyed
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp=reduced_coders_df.loc[reduced_coders_df["AttendedBootcamp"]=="Yes"]
```

```python
# Calculate average age of attendees

average_age=attended_bootcamp["Age"].mean()


# Calculate how many people attended a bootcamp
number_of_attendees=attended_bootcamp["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees

holds_degree=attended_bootcamp["SchoolDegree"].count()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender=attended_bootcamp["Gender"].count()
male=attended_bootcamp["Gender"].value_counts()['male']
female=attended_bootcamp["Gender"].value_counts()['female']
non_specific= total_gender-male-female

# Calculate percentage of respondents who attended a bootcamp

attended_bootcamp_percentage=(number_of_attendees/total_surveyed)*100

# Calculate percentage of respondents belonging to each gender
male_percent=(male/total_surveyed)*100
female_percent=(female/total_surveyed)*100
non_specific_percent=(non_specific/total_surveyed)*100

# Calculate percentage of attendees with a school degree

holds_degree_percent=(holds_degree/number_of_attendees)*100

# Calculate average post-bootcamp salary
avg_salary=attended_bootcamp["BootcampPostSalary"].mean()
```

```python
# Create a new table consolodating above calculations
summary_table=pd.DataFrame({"Total Surveyed":[total_surveyed],
                                           "Total Bootcamp Attendees":[number_of_attendees],
                                           "Percent Attended a Bootcamp":[attended_bootcamp_percentage],
                                           "Average Age":[average_age],
                                           "% Male":[male_percent],
                                           "% Female":[female_percent],
                                           "% Non Gender Specific":[non_specific_percent],
                                           "% with a degree":[holds_degree_percent],
                                           "Average Post Bootcamp Salary":[avg_salary]})

summary_table=summary_table[["Total Surveyed", "Total Bootcamp Attendees", "Average Age", "% Male", "% Female", "% Non Gender Specific","% with a degree","Average Post Bootcamp Salary"]]
summary_table=summary_table.round(2)
summary_table
```

```python
# Improve formatting before outputting spreadsheet

summary_table["% Male"]=summary_table["% Male"].map("{0:,.2f}%".format)
summary_table["% Female"]=summary_table["% Female"].map("{0:,.2f}%".format)
summary_table["Percent Attended a Bootcamp"] = summary_table["Percent Attended a Bootcamp"].map("{0:,.2f}%".format)
summary_table["% Non Gender Specific"] = summary_table["% Non Gender Specific"].map("{0:,.2f}%".format)
summary_table["% with a degree"]= summary_table["% with a degree"].map("{0:,.2f}%".format)
summary_table["Average Post Bootcamp Salary"]= summary_table["Average Post Bootcamp Salary"].map("${0:,0f}".format)
summary_table
```

```python
# Export to Excel

summary_table.to_excel("bootcamppart1.xlsx", index=False)
```
